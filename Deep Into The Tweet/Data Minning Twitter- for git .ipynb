```
def get_tweets_for_loops(keyword,topic):

  filters = '-filter:media'



  
  search = tweepy.Cursor(api.search, q =  keyword+filters, 
                      tweet_mode = 'extended', 
                      include_rts = True,
                      lang = "en").items(100)
                      
                      
                      
# Create lists for each field desired from the tweets
  tweet_data = {'tweet_id':[], 'tweet_created_at':[], 'source':[], 'original_text':[], 'tweet_favorite_count':[], 
                'retweet_count':[], 'hashtags':[], 'user_mentions':[],'user_mentions_id':[], 'topic':[], 
                'keyword':[], 'emojis':[], 'possibly_sensitive':[],
                'in_reply_to_status_id':[], 'in_reply_to_user_id':[], 'in_reply_to_screen_name':[],
                'user_id':[], 'user_followers_count':[], 'user_listed_count':[], 'user_favourites_count':[],
                'user_statuses_count':[], 'user_created_at':[], 'user_screen_name':[],'place':[],
                'latitude_corr':[], 'longitude_corr':[], 'place_coord_boundaries':[],'retweet_tweet_id':[],
                'retweet_user_id':[], 'retweet_user_screen_name':[], 'retweet_in_reply_to_screen_name':[],
                'retweet_in_reply_to_status_id':[],'retweet_in_reply_to_user_id':[]}




  for tweet in search:
    
    tweet_data['topic'].append(topic)
    tweet_data['keyword'].append(keyword)
    
    status = tweet
    if 'extended_tweet' in status._json: status_json = status._json['extended_tweet']['full_text']
    elif 'retweeted_status' in status._json and 'extended_tweet' in status._json['retweeted_status']:
       status_json = status._json['retweeted_status']['extended_tweet']['full_text'] 
    elif 'retweeted_status' in status._json: status_json = status._json['retweeted_status']['full_text']
    else: status_json = status._json['full_text']
    status = status._json
    tweet_data['original_text'].append(status_json)

    tweet_data['tweet_id'].append(status['id'])
    tweet_data['tweet_created_at'].append(status['created_at'])
    tweet_data['source'].append(status['source'])
    tweet_data['tweet_favorite_count'].append(status['favorite_count'])
    tweet_data['retweet_count'].append(status['retweet_count'])

    tweet_data['emojis'].append(demoji.findall(status_json))

    mentions = status["entities"]["user_mentions"]
    mentioned_users  = ", ".join([mention['screen_name'] for mention in mentions])
    tweet_data['user_mentions'].append(mentioned_users ) 
    mentions_id  = [mention['id'] for mention in mentions]
    tweet_data['user_mentions_id'].append(mentions_id)

    hashtags = ", ".join([hashtag_item['text'] for hashtag_item in status['entities']['hashtags']])
    tweet_data['hashtags'].append(hashtags)

    tweet_data['in_reply_to_status_id'].append(tweet.in_reply_to_status_id)
    tweet_data['in_reply_to_user_id'].append(tweet.in_reply_to_user_id)
    tweet_data['in_reply_to_screen_name'].append(tweet.in_reply_to_screen_name)

    tweet_data['user_screen_name'].append(status['user']['screen_name'])#
    tweet_data['user_id'].append(status['user']['id'])
    tweet_data['user_followers_count'].append(status['user']['followers_count'])
    tweet_data['user_listed_count'].append(status['user']['listed_count'])
    tweet_data['user_favourites_count'].append(status['user']['favourites_count'])
    tweet_data['user_statuses_count'].append(status['user']['statuses_count'])
    tweet_data['user_created_at'].append(status['user']['created_at'])


        #get retweets data of the tweet if possible
    if "retweeted_status" in status:
      tweet_data['retweet_tweet_id'].append(tweet.retweeted_status.id)
      tweet_data['retweet_user_id'].append(tweet.retweeted_status.user.id)
      tweet_data['retweet_user_screen_name'].append(tweet.retweeted_status.user.screen_name)
      tweet_data['retweet_in_reply_to_screen_name'].append(tweet.retweeted_status.id)
      tweet_data['retweet_in_reply_to_status_id'].append(tweet.retweeted_status.in_reply_to_status_id)
      tweet_data['retweet_in_reply_to_user_id'].append(tweet.retweeted_status.in_reply_to_user_id)
    else:
      tweet_data['retweet_tweet_id'].append(None)
      tweet_data['retweet_user_id'].append(None)
      tweet_data['retweet_user_screen_name'].append(None)
      tweet_data['retweet_in_reply_to_screen_name'].append(None)
      tweet_data['retweet_in_reply_to_status_id'].append(None)
      tweet_data['retweet_in_reply_to_user_id'].append(None)


    #get location of the tweet if possible
    try:
        location = status['user']['location']
        user_loc = geolocator.geocode(location)

        if user_loc:              
          
          lat = user_loc.latitude
          lon = user_loc.longitude
        else:
          lat = ''
          lon = ''                               

    except TypeError:
        location = ''
      
    tweet_data['latitude_corr'].append(lat)
    tweet_data['longitude_corr'].append(lon)
    tweet_data['place'].append(location)

    try:
        coordinates = [coord for loc in status['place']['bounding_box']['coordinates'] for coord in loc]
    except TypeError:
        coordinates = None
    tweet_data['place_coord_boundaries'].append(coordinates)


#Is the tweet possibly sensitive?
    try:
      is_sensitive = status['possibly_sensitive']
    except KeyError:
      is_sensitive = None
    tweet_data['possibly_sensitive'].append(is_sensitive)


  tweets_df = pd.DataFrame(tweet_data)
  time.sleep(1)

  return (tweets_df)
  
```

```
words = { 'wikipedia':'Technology','Linux':'Technology','dog':'animals'}
```

```
d = words
keywords = list(d.keys())
topics = list(d.values())
tweets_topics = []
for i in range(len(keywords)):
  tweets_topics.append(get_tweets_for_loops(keywords[i],topics[i]))
  print(keywords[i])
tweets_topics = pd.concat(tweets_topics)

```
